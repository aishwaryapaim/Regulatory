```python
df = pd.read_excel('Y:\\delete\\European Commission\\ActiveSubstances.xls',skiprows=2)
df.columns.values[2] = "approved"
df = df[pd.to_numeric(df['Substance'], errors='coerce').isna()]
df.dropna(inplace = True)
mylist = df['Substance'].to_numpy()
approved = df['approved'].to_numpy()
mylist.shape
```

```python
cas = np.array([])
x=0
p = re.compile("[0-9]+-[0-9]+-[0-9]+")
l2 = []
for i in mylist:
    print(i)
    base_url = 'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/'+str(i)+'/synonyms/txt'
    r = requests.get(base_url)
    data = r.text
    lst = data.split('\n')
    if lst[0] != 'Status: 404' :
        l2 = [ s for s in lst if p.match(s) ]
        if len(l2)>0 :
                    l2 = np.array([l2[0]])        
        else:
            l2 = ['NA']
    else:
        if i.find('(') > 1:
            x = i.split(' (')
            base_url = 'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/'+str(x[0])+'/synonyms/txt'
            r = requests.get(base_url)
            data = r.text 
            lst = data.split('\n')
            if lst[0] != 'Status: 404' :
                l2 = [ s for s in lst if p.match(s) ]
                if len(l2)>0 :
                    l2 = np.array([l2[0]])        
                else:
                    l2 = ['NA']
            else:
                l2 = ['NA']
        else:
                l2 = ['NA']
    print(l2)            
    cas = np.concatenate((cas, l2))
```

```python
dataset = pd.DataFrame({'mylist':mylist,'CAS Number':cas,'approved':approved},columns=['mylist','CAS Number','approved'])
dataset = dataset.loc[dataset['CAS Number'] != 'NA']
dataset.dropna(inplace = True)
```

```python
CAS = pd.read_excel('Y:/delete/Chemical CAS.xlsx')
CAS['value'] = 'Approved'
CAS = CAS.applymap(lambda x: x.strip() if isinstance(x, str) else x)
result = pd.merge(dataset, CAS, how="outer", on="CAS Number")
result.approved.fillna(result.value, inplace=True)
del result['value']
```

```python
result['Jurisdiction'] = 'Europe'
result['RegulatoryAgency'] = 'European Commission'
result['TypeOfRegulation'] = 'Use'
result['Criteria'] = 'Pesticide approval status'

result['Description'] = 'Pending or N/a'
result.loc[result['approved'] == 'Not approved', 'Description'] = 'Not Approved'
result.loc[result['approved'] == 'Approved', 'Description'] = 'Approved'


result['HazardClassification'] = 'Moderate'
result.loc[result['approved'] == 'Not approved', 'HazardClassification'] = 'High'
result.loc[result['approved'] == 'Approved', 'HazardClassification'] = 'Low'

result['CauseOfClassification'] = 'Pending, not PPP (i.e., not yet assessed at the EU level)'
result.loc[result['approved'] == 'Not approved', 'CauseOfClassification'] = 'Not approved for use as a pesticide'
result.loc[result['approved'] == 'Approved', 'CauseOfClassification'] = 'Approved for use as a pesticide or N/A'


result['SourceName'] = 'European Commission'
result['SourceURL'] = 'https://ec.europa.eu/food/plant/pesticides/eu-pesticides-database/active-substances/?event=search.as'

result = result.drop(['mylist', 'approved'], axis=1)

result.head(2)
```

```python
results.to_csv('Y:/delete/Reugulatory Assessment.csv',index=False,mode='a',header=False)
```
